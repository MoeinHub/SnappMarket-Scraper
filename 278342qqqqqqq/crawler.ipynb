```python
import json
import requests
def paginator(total):
    if total<200:
        return 1
    else:
        return (total-1)//200+1
    
def reqer(cat_id=278342, offset=0, limit=0, vendor="0r5ryz"):
    addr=f'https://core.snapp.market/api/v2/vendors/{vendor}/products?limit={limit}&offset={offset}&categories[]={cat_id}'
    res=requests.get(addr)
    return res, res.json()

def get_prods(cat_id=None):
    offset=0
    res, jres = reqer(cat_id)
    res_total=jres['metadata']['pagination']['total']
    pages=paginator(res_total)
    page=0
    while page<pages:
        if offset<=3000:
            res, jres=reqer(cat_id, offset)
            if res.status_code==200:
                print(jres)
            page+=1
            offset+=200
        else:
            print("offset 3000 limit reached !")
            break

def subcats(id=None):
    req, jreq=reqer(id) #for test root (development)
    #req, jreq=basereq, jbasereq #for production and sending requests
    base=jreq['metadata']['filter']['options'][0]['data']['options']
    catbrands=jreq['metadata']['filter']['options'][1]['data']['options']
    if len(base)!=1:# or base[0]['id'] != id:#len base root is more than 1
        res1=base
    else:
        while base[0]['id'] != id:
            base=base[0]['children']
        res1=base
        res1=base[0]['children']    
    return res1, catbrands
# subcats()

# def's
```
